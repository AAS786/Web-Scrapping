<a id="1"></a>  
#### <p style="padding:8px;background-color:#d5eae9;margin:0;color:#c27849;font-family:newtimeroman;font-size:150%;text-align:left;border-radius: 70px 60px;overflow:hidden;font-weight:200">Normal Distribution </p>

```python
# Assuming df is your DataFrame
data = df['Stars']

# Calculate statistics
mean_val = data.mean()
median_val = data.median()
skewness_val = skew(data)
kurtosis_val = kurtosis(data)



# Create a normal distribution curve
sns.distplot(data, fit=norm, kde=False)
plt.title('Normal Distribution Curve')
plt.xlabel(' Stars')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Print statistics
print(f'Mean: {mean_val}')
print(f'Median: {median_val}')
print(f'Skewness: {skewness_val}')
print(f'Kurtosis: {kurtosis_val}')
```

<a id="1"></a>  
#### <p style="padding:8px;background-color:#d5eae9;margin:0;color:#c27849;font-family:newtimeroman;font-size:150%;text-align:left;border-radius: 70px 60px;overflow:hidden;font-weight:200">Mean & Median </p>

```python

# Assuming df is your DataFrame
data = df['Stars']

# Create a histogram with a normal distribution fit
sns.distplot(data, fit=norm, kde=False)

# Calculate mean and median
mean_val = np.mean(data)
median_val = np.median(data)

# Plot mean and median lines
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_val}')

# Set title and labels
plt.title('Normal Distribution Curve with Mean and Median Lines')
plt.xlabel('Review Stars')
plt.ylabel('Frequency')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

<a id="1"></a>  
#### <p style="padding:8px;background-color:#d5eae9;margin:0;color:#c27849;font-family:newtimeroman;font-size:150%;text-align:left;border-radius: 70px 60px;overflow:hidden;font-weight:200">Skewness</p>

```python

# Assuming df is your DataFrame
data = df['Stars']

# Create a histogram with a normal distribution fit
sns.distplot(data, fit=norm, kde=False)

# Calculate mean, median, and skewness
mean_val = np.mean(data)
median_val = np.median(data)
skewness_val = data.skew()

# Plot mean and median lines
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_val}')

# Add text annotations for skewness
if skewness_val > 0:
    plt.text(plt.xlim()[1], 0.015, f'Positive Skewness: {skewness_val:.2f}', horizontalalignment='right', verticalalignment='bottom')
elif skewness_val < 0:
    plt.text(plt.xlim()[0], 0.015, f'Negative Skewness: {skewness_val:.2f}', horizontalalignment='left', verticalalignment='bottom')

# Set title and labels
plt.title('Normal Distribution Curve with Mean, Median, and Skewness Annotations')
plt.xlabel('Review Stars')
plt.ylabel('Frequency')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python

# Assuming df is your DataFrame
data = df['Stars']

# Create a histogram with a normal distribution fit
sns.distplot(data, fit=norm, kde=False)

# Calculate mean, median, skewness, and kurtosis
mean_val = np.mean(data)
median_val = np.median(data)
skewness_val = skew(data)
kurtosis_val = kurtosis(data)

# Plot mean, median, skewness, and kurtosis lines
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_val}')
plt.axvline(mean_val + skewness_val, color='purple', linestyle='dashed', linewidth=2, label=f'Skewness: {skewness_val:.2f}')
plt.axvline(mean_val + kurtosis_val, color='orange', linestyle='dashed', linewidth=2, label=f'Kurtosis: {kurtosis_val:.2f}')

# Set title and labels
plt.title('Normal Distribution Curve with Mean, Median, Skewness, and Kurtosis Lines')
plt.xlabel('Review Stars')
plt.ylabel('Frequency')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
